<h3>Z-Index Standardisation for Continuous Variables</h3>

```python
#columns with continuous variable
cols = list(df.columns[12:-1])
cols.append("LIMIT_BAL")
cols.append("AGE")
cols
```

```python
#Standardise Function
def standardise(df, columns):
    new_df = df.copy(deep=True)
    for col in columns:
        new_df[col] = (df[col] - df[col].mean()) / df[col].std()
    return new_df
```

```python
#Model Evaluation
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, balanced_accuracy_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
from imblearn.metrics import specificity_score

eval_methods={accuracy_score:"Accuracy", recall_score:"Recall", precision_score:"Precision", specificity_score:"Specificity", balanced_accuracy_score:"Average Class Accuracy", roc_auc_score:"ROC Index", f1_score:"F1 score"}
```

```python
#Stratified K-Fold Cross-Validation
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=10)
feature_cols = df4.columns[1:-1]
features = df4[feature_cols]
target_col = df4.columns[-1]
target = df4[target_col]
for model in models:
    for e in eval_methods.keys():
        total = 0
        cm = [[0,0],[0,0]]
        for train_index, test_index in skf.split(df4, df4['default payment next month']):
            train = df4.loc[train_index,:]
            test = df4.loc[test_index,:]
            X_train, y_train = train[feature_cols], train[target_col]
            X_test, y_test = test[feature_cols], test[target_col]
            trained_model = model.fit(X_train, y_train)
            y_pred = trained_model.predict(X_test)
            total += e(y_test, y_pred)
            cm += confusion_matrix(y_test, y_pred)
        print(type(model).__name__,"-",eval_methods[e],":", total/10)
    cm_display = ConfusionMatrixDisplay(cm//10, display_labels = [False, True])
    cm_display.plot()
    plt.show()
```
